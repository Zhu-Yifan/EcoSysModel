# Single Species Growth Model

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
fig_dir = "E:\OneDrive\EcoSysModel\plot"
os.makedirs(fig_dir, exist_ok=True)
```

## Logistic Growth Model
###  with density-dependent control, population (N) will stabilize at carrying capacity

```python
# compute single species growth curve
N0 = 5 # initial population
r = 0.5 # specific growth rate
K = 60 # carrying capacity
t = np.arange(0, 51, 0.5)

N = K / ( 1 + (K/N0 -1)* np.exp(-r*t) )

# plot
plt.figure()
plt.plot(t, N, marker = 'o', color = 'blue')
plt.xlabel('Time (days)', fontsize = 15)
plt.ylabel('Population (individuals)', fontsize = 15)
plt.tick_params(axis='both', labelsize=12)

# save plot
plt.savefig(f'{fig_dir}/single_species_Logistic_Growth_Model.png', bbox_inches='tight', dpi = 300)
```

Note:

* Using different time steps based on numerical solutions (an Euler approximation)

* the finite difference method becomes a better approximation for small Δt

* In both numerical and analytical solutions, the population stabilizes at K, but the rate of stabilization and how quickly the population gets close to K depends on the time step Δt.


```python
plt.figure(figsize=(15, 5))

steps = [0.1, 1.0, 2.0] # Δt: 0.1 day, 1 day, and 2 day
N0 = 5 # initial population
r = 0.5 # specific growth rate
K = 60 # carrying capacity

c = ['blue','green','red']
ind = [1,2,3]
for dt,idx in zip(steps, ind):
    print(dt,idx)
    N = [N0]
    t = np.arange(0, 50 + dt, dt)  # Generate time array based on time step

    # Calculate population using the finite difference method
    for _ in range(len(t) - 1):
        Nt = N[-1] + dt * r * N[-1] * (1 - N[-1] / K)  # Population growth equation
        N.append(Nt)

    plt.subplot(1,3,idx)
    plt.plot(t, N, marker = 'o', color = c[idx-1])
    plt.xlabel('Time (days)', fontsize = 15)
    plt.ylabel('Population (individuals)', fontsize = 15)
    plt.title(f'time step = {dt} day', fontsize = 15)
    plt.tick_params(axis='both', labelsize=12)
plt.savefig(f'{fig_dir}/single_species_logistic_growth.png', bbox_inches='tight', dpi = 300)
```

## Growth Model with unlimited resources
*  population (N) will grow following the power law
* Thomas Malthus, 1798 “An Essay on the Principle of Population“


```python
# compute single species growth curve

t = np.arange(0, 51, 0.5)
N = np.exp(0.5*t)

# plot
plt.figure()
plt.plot(t, N, marker = 'o', color = 'blue')
plt.xlabel('Time (days)', fontsize = 15)
plt.ylabel('Population (individuals)', fontsize = 15)
plt.tick_params(axis='both', labelsize=12)

# save plot
plt.savefig(f'{fig_dir}/single_species_power_law.png', bbox_inches='tight', dpi = 300)
```

Note:
* Using different time steps based on numerical solutions (an Euler approximation)
* Different time steps led to difference in orders of magnitude, largely deviating from the analytic solution.

```python
# create for loop
plt.figure(figsize=(15, 5))

steps = [0.1, 1.0, 2.0] # Δt: 0.1 day, 1 day, and 2 day
N0 = 5 # initial population
r = 0.5 # specific growth rate， unit: day^(-1)

c = ['blue','green','red']
ind = [1,2,3]

for dt,idx in zip(steps, ind):
    print(dt,idx)
    N = [N0]
    t = np.arange(0, 50 + dt, dt)  # Generate time array based on time step

    # Calculate population using the finite difference method
    for _ in range(len(t) - 1):
        Nt = N[-1] + dt * r * N[-1]  # Population growth equation 
        N.append(Nt)
    
    plt.subplot(1,3,idx)
    plt.plot(t, N, marker = 'o', color = c[idx-1])
    plt.xlabel('Time (days)', fontsize = 15)
    plt.ylabel('Population (individuals)', fontsize = 15)
    plt.title(f'time step = {dt} day', fontsize = 15)
    plt.tick_params(axis='both', labelsize=12)
plt.savefig(f'{fig_dir}/single_species_unlimited_resource.png', bbox_inches='tight', dpi = 300)        
```
